```python
# Hack to remove pandas FutureWarning
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import os
import gensim
import pickle

# Load the stuff we saved when running LDA
dictionary = gensim.corpora.Dictionary.load(os.path.join('output', 'dictionary.gensim'))
corpus = pickle.load(open(os.path.join('output', 'corpus.pkl'), "rb"))
lda = gensim.models.ldamodel.LdaModel.load(os.path.join('output', 'model5.gensim'))
```

```python
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
```

```python
lda_display = gensimvis.prepare(lda, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
```

```python
lda3 = gensim.models.ldamodel.LdaModel.load(os.path.join('output', 'model3.gensim'))
lda_display3 = gensimvis.prepare(lda3, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display3)
```
